## 1.How many are there in each Team and the percentage splitting with respect to the total employees.


```python
team_count=emp_df['Team'].value_counts()
team_count
```

```python
plt.figure(figsize=(30,6))
sns.countplot(x='Team',data=emp_df,hue='Team',palette='viridis')
plt.xticks(rotation="vertical")
plt.show()
```

```python
team_df=pd.DataFrame({'Team':team_count.index,'Count':team_count.values})
total_mem=team_df['Count'].sum()
Percentage=round((team_df['Count']/total_mem)*100,2)
team_df['Percentage splitting']=Percentage
team_df
```

## INFERENCE

In conclusion, the company exhibits a diverse distribution of teams with a total of 30 distinct teams. The team counts range from 14 to 19.The New Orleans Pelicans and Memphis Grizzlies stand out with team counts of 19 and 18, respectively.The lowest counts observed for the Orlando Magic and Minnesota Timberwolves, each having 14 members.
The team percentage splitting ranges from 3.06% to 4.15%.New Orleans Pelicans has the highest percentage splitting with 4.15%, while Orlando Magic and Minnesota Timberwolves have the lowest with 3.06%.

## 4.Find out under which team and position, spending in terms of salary is high.

```python
salary_by_position=emp_df.groupby(['Position'])['Salary'].sum()
salary_by_position
```

```python
print("Position with highest salary:",salary_by_position.idxmax())
```

```python
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")
sns.barplot(x='Position', y='Salary', data=emp_df, estimator=sum, ci=None, palette='Blues_d')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.title('Total Salary by Position')
```

```python
salary_by_team=emp_df.groupby(['Team'])['Salary'].sum()
salary_by_team
```

```python
print("Team with highest salary:",salary_by_team.idxmax())
```

```python
plt.figure(figsize=(12, 6))
plt.xticks(rotation=90)
sns.set_theme(style="whitegrid")
sns.barplot(x='Team', y='Salary', data=emp_df, estimator=sum, ci=None, palette='Blues_d')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.title('Total Salary by Team')
```

## 5.Find if there is any correlation between age and salary , represent it visually.

```python
df2=emp_df[['Age','Salary']]
df2
```

```python
corr = df2.corr()
print(corr)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix between Age and Salary")
plt.show()
```

## INFERENCE
The correlation matrix indicates that there's a positive but not particularly strong relationship between age and salary in the employee dataset since the value of correlation coefficient is only .21. As employees' ages tend to increase, there is a tendency for their salaries to slightly rise.
